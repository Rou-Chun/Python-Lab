```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import tkinter as tk
```

```python
window = tk.Tk()
window.title('比大小賭博遊戲')
window.geometry('500x500')

initialStr = tk.StringVar()
initial = np.random.randint(100)

numStr = tk.StringVar()
num = np.random.randint(100)

resultText = tk.StringVar()

count = 0
wincount = 0
money = 5

def playBig():
    global initial, num ,money, count, wincount
    count+=1
    if money > 0:
        if initial < num:
            wincount+=1
            money+=1
            resultText.set('You Win!\nYou still have '+str(money)+' dollars')        
        else:
            money-=1
            resultText.set('You Lose!\nYou still have '+str(money)+' dollars')        
        
        initialStr.set(str(initial))
        numStr.set(str(num)) 
    
        if money == 0:
            resultText.set('Game over!\nWin rate: '+str(round((wincount/count)*100,2))+'%')
        else:           
            initial = np.random.randint(100)
            num = np.random.randint(100)
def playSmall():
    global initial, num ,money, count, wincount
    count+=1
    if money > 0:
        if initial > num:
            wincount+=1
            money+=1
            resultText.set('You Win!\nYou still have '+str(money)+' dollars')        
        else:
            money-=1
            resultText.set('You Lose!\nYou still have '+str(money)+' dollars')            
    
        initialStr.set(str(initial))    
        numStr.set(str(num))
        
        if money == 0:
            resultText.set('Game over!\nWin rate: '+str(round((wincount/count)*100,2))+'%')
        else:            
            initial = np.random.randint(100)
            num = np.random.randint(100)

b1 = tk.Button(window, text = '押大', width = 10, height = 5, command = playBig)
b2 = tk.Button(window, text = '押小', width = 10, height = 5, command = playSmall)

b1.place(x=120,y=50)
b2.place(x=300,y=50)

l1 = tk.Label(window, text = 'Compare number：', bg = '#00E600',width = 15, height = 2)
l1.place(x=105,y=150)
compare = tk.Label(window, textvariable = initialStr, bg = '#00E600',width = 15, height = 2)
compare.place(x=105,y=200)

l2 = tk.Label(window, text = 'Your number：', bg = '#00E6E6',width = 15, height = 2)
l2.place(x=285,y=150)
you = tk.Label(window, textvariable = numStr, bg = '#00E6E6',width = 15, height = 2)
you.place(x=285,y=200)

result = tk.Label(window, textvariable = resultText, bg = '#FFAC12',width = 20, height = 4)
result.place(x=200,y=300)

window.mainloop()
```
