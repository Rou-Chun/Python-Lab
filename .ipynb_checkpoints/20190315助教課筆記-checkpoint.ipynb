# 畫圖技巧複習

```python
x = np.linspace(0,10,1000)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
plt.figure(figsize=(6,6)) #設定圖的大小
plt.xkcd() #設定圖的風格

plt.plot(x,y1,label = 'sin(x)')
plt.plot(x,y2,label = 'cos(x)')
plt.legend('lower left') #設定圖例位置

plt.xlabel("X Axis") #設定x軸名稱
plt.ylabel("Y Axis") #設定y軸名稱
plt.xlim(0,10) #設定x軸範圍
plt.ylim(-1,1) #設定y軸範圍

plt.xticks(np.arange(0,10,1)) #x軸分成10部分
plt.yticks(np.arange(-1,1,0.5)) #y軸分成4部分
plt.title('Sin and Cos') #設定圖的標題
```

# 加*讓變數長度不受限

```python
def ARGV(*argv):
    for i in argv:
        print(i)
```

```python
ARGV("H","W","I","A","M" )
```

```python
def ARGV1(arg1, *arg):
    print(arg1)
    
    print("")
    for i in arg:
        print(i)
```

```python
ARGV1("H","W","I","A","M" )
```

```python
def ARGV2(*arg, arg1):
    print(arg1)
    
    print("")
    for i in arg:
        print(i)
```

```python
ARGV2("H","W","I","A","M",arg1="Y")
```

# 加**讓變數讀字典形式

```python
def KWARG(**kwargs):
    for key,value in kwargs.items():
        print(f'{key} == {value}')
```

```python
KWARG(first = "N",second = "C", third = "T", fourth = "S")
```

```python
def Four(arg1,arg2,arg3,arg4):
    print("arg1 is", arg1)
    print("arg2 is", arg2)
    print("arg3 is", arg3
    print("arg4 is", arg4)
```

```python
args = ("N","C","T","S")
```

```python
Four(*args)
```

```python
kwargs = {"arg2":"T","arg1":"N","arg4":"S","arg3":"C"}
```

```python
Four(**kwargs)
```

# Enumerate基本介紹

```python
mylist = ["N","C","T","S"]
```

```python
mylist2 = list(enumerate(mylist))
mylist2
```

```python
mylist3 = dict(enumerate(mylist))
mylist3
```

```python
mylist4 = list(enumerate(mylist, 7))
mylist4
```

# map and lambda 基本介紹

```python
def square(x):
    return x**2
```

```python
list(map(square,[1,2,3,4,5]))
```

lambda定義的函數只在該行有效

```python
list(map(lambda x:x**2,[1,2,3,4,5]))
```
