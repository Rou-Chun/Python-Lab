```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
r1 = 3.0
a1, b1 = (0., 0.)
theta = np.arange(0, 2*np.pi, 0.01)
x1 = a1 + r1 * np.cos(theta)
y1 = b1 + r1 * np.sin(theta)
r2 = 1.0
a2, b2 = (0., -1.)
x2 = a2 + 1.5 * r2 * np.cos(theta)
y2 = b2 + r2 * np.sin(theta)
r3 = 0.5
a3, b3 = (-1.3, 1.2)
x3 = a3 + r3 * np.cos(theta)
y3 = b3 + r3 * np.sin(theta)
r4 = 0.5
a4, b4 = (1.3, 1.2)
x4 = a4 + r4 * np.cos(theta)
y4 = b4 + r4 * np.sin(theta)
r5 = 0.3
a5, b5 = (-0.5, -1.)
x5 = a5 + r5 * np.cos(theta)
y5 = b5 + r5 * np.sin(theta)
r6 = 0.3
a6, b6 = (0.5, -1.)
x6 = a6 + r6 * np.cos(theta)
y6 = b6 + r6 * np.sin(theta)
x7=[-2.5, -2, -1.5]
y7=[1.7, 3.5, 2.65]
x8=[2.5, 2, 1.5]
y8=[1.7, 3.5, 2.65]
```

```python
plt.plot(x1,y1)
plt.plot(x2,y2)
plt.plot(x3,y3)
plt.plot(x4,y4)
plt.scatter(a3,b3)
plt.scatter(a4,b4)
plt.plot(x5,y5)
plt.plot(x6,y6)
plt.plot(x7,y7)
plt.plot(x8,y8)
```
