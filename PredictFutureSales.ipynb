先複製一份原始資料，將日期格式分成日、月、年合併到data_sales_copy裡

input檔案格式解釋：

第一列：
shop_id = 59, item_id = 22154 在 2013 年 1 月，每一天的銷售量[0,0,0,....,0] (長度為31) (1月沒有賣出)


第二列：
shop_id = 59, item_id = 22154 在 2013 年 2 月，每一天的銷售量[1,0,0,....,0] (長度為31) (只有2/1有賣出)
...
...


以此類推，每一間商店每一項商品都會有34列的資料(表示2013年1月到2015年10月)

```python
#2013年shop_item[0]的組合

everydaySalesInMonth = np.zeros(31)
day = data_sales_copy.loc[(data_sales_copy['Year'] == 2013) & (data_sales_copy['Month'] == 1) & (data_sales_copy['shop_id'] == shop_item.loc[0][0]) & (data_sales_copy['item_id'] == shop_item.loc[0][1]),'day']
everydaySalesInMonth[day - 1] = data_sales_copy.loc[(data_sales_copy['Year'] == 2013) & (data_sales_copy['Month'] == 1) & (data_sales_copy['shop_id'] == shop_item.loc[0][0]) & (data_sales_copy['item_id'] == shop_item.loc[0][1]),'item_cnt_day']
X_train = everydaySalesInMonth.reshape(1,31)
X_train = pd.DataFrame(X_train)



for i in range(2,13):
    everydaySalesInMonth = np.zeros(31)
    day = data_sales_copy.loc[(data_sales_copy['Year'] == 2013) & (data_sales_copy['Month'] == i) & (data_sales_copy['shop_id'] == shop_item.loc[0][0]) & (data_sales_copy['item_id'] == shop_item.loc[0][1]),'day']
    everydaySalesInMonth[day - 1] = data_sales_copy.loc[(data_sales_copy['Year'] == 2013) & (data_sales_copy['Month'] == i) & (data_sales_copy['shop_id'] == shop_item.loc[0][0]) & (data_sales_copy['item_id'] == shop_item.loc[0][1]),'item_cnt_day']
    everydaySalesInMonth = everydaySalesInMonth.reshape(1,31)
    everydaySalesInMonth = pd.DataFrame(everydaySalesInMonth)
    X_train = pd.concat([X_train, pd.DataFrame(everydaySalesInMonth)],axis=0)
```

```python
X_train
```

```python
x_train = X_train.values
#y_train = Y_train.values
y_train = [1,2,3,4,5,5,4,3,2,1,0,0]
```

```python
from keras.preprocessing import sequence
from keras.models import Sequential
from keras.layers import Dense,Embedding, LSTM, Dropout
```

```python
x_element_MAX_Dimention = 2169+22
N=10 #2169+22要壓到幾維?
K=10 #LSTM的個數
```

```python
model = Sequential()
```

```python
model.add(Embedding(x_element_MAX_Dimention,N)) 
model.add(LSTM(K))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='mse', optimizer='adam', metrics=['accuracy'])
model.summary()
```

```python
model.fit(x_train, y_train, batch_size=100, epochs=2)
```
